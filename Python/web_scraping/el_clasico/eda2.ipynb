### **Replacing non-standard characters**


```python
file = pd.read_csv("Datasets/sorted_df.csv")
df = pd.DataFrame(file)
```

```python
def clean_text(text):
    if isinstance(text, str):
        # Split by commas
        names = text.split(',')
        
        # Clean each name
        cleaned_names = []
        for name in names:
            # Remove leading/trailing whitespace
            name = name.strip()
            
            # Additional specific replacements
            replacements = {
                'Ã¡': 'a',
                'Ã­': 'i',
                'Ã³': 'o',
                'Ã ': 'a',
                'Ã¢': 'a',
                'Ã±': 'n',
                'Ã': 'i',
                'Â©': 'e',
                'Ã': 'A',
                '©': 'e',
                'Å': 'n',
                'Ã': 'o',
                'Ã': 'a'
            }
            
            for old, new in replacements.items():
                name = name.replace(old, new)
            
            # Normalize Unicode characters
            normalized = unicodedata.normalize('NFKD', name)
            # Remove diacritics (keeping base characters)
            cleaned = ''.join(c for c in normalized if not unicodedata.combining(c))
            cleaned_names.append(cleaned)
        
        # Join back with commas and spaces
        return ', '.join(cleaned_names)
    return text

 

```

```python

# Clean both columns
df['Barcelona_Scorers'] = df['Barcelona_Scorers'].apply(clean_text)
df['Real_Madrid_Scorers'] = df['Real_Madrid_Scorers'].apply(clean_text)
```

```python
df['Real_Madrid_Own_Goals'] = df['Real_Madrid_Own_Goals'].apply(clean_text)
df['Barcelona_Own_Goals'] = df['Barcelona_Own_Goals'].apply(clean_text)
df['Venue'] = df['Venue'].apply(clean_text)
```

```python
df.to_csv("Datasets/clean_df.csv", index=False)
```

```python
def replace_ronaldo(text):
    if isinstance(text, str):
        # Split by comma
        names = [name.strip() for name in text.split(',')]
        
        # Replace each instance of 'Ronaldo'
        cleaned_names = ['Valle Mas' if name == 'Mas' else name for name in names]
        
        # Join back with commas
        return ', '.join(cleaned_names)
    return text
```

```python
df['Barcelona_Scorers'] = df['Barcelona_Scorers'].apply(replace_ronaldo)
```

```python
def replace_suarez(row):
    if pd.notna(row['Barcelona_Scorers']) and 2002 <= row['Year'] <= 2007:
        # Split names by comma
        names = [name.strip() for name in row['Barcelona_Scorers'].split(',')]
        
        # Replace Suarez with L.A.Suarez
        names = ['Ronaldo Nazario' if name.strip() == 'Ronaldo' else name for name in names]
        
        # Join back with commas
        return ', '.join(names)
    return row['Barcelona_Scorers']
```

```python
df['Real_Madrid_Scorers'] = df.apply(replace_suarez, axis=1)
```

```python
def count_scorers(scorers_str):
    if pd.isna(scorers_str) or scorers_str == '':
        return 0
    return len(scorers_str.split(','))

# Calculate goals from scorers and own goals for Barcelona
df['Calculated_Barcelona_Goals'] = df.apply(lambda row: 
    count_scorers(row['Barcelona_Scorers']) + 
    (0 if pd.isna(row['Real_Madrid_Own_Goals']) else count_scorers(row['Real_Madrid_Own_Goals'])), 
    axis=1)

# Find mismatches
mismatches = df[df['Barcelona_Goals'] != df['Calculated_Barcelona_Goals']]

# Print mismatches with relevant columns
if len(mismatches) > 0:
    print("Found mismatches in the following matches:")
    for _, row in mismatches.iterrows():
        print(f"\nDate: {row['Date']}")
        print(f"Recorded Barcelona Goals: {row['Barcelona_Goals']}")
        print(f"Barcelona Scorers: {row['Barcelona_Scorers']}")
        print(f"Real Madrid Own Goals: {row['Real_Madrid_Own_Goals']}")
        print(f"Calculated Goals: {row['Calculated_Barcelona_Goals']}")
else:
    print("No mismatches found - all goal counts are consistent!")
```

```python
def verify_winner(row):
    barcelona_goals = row['Barcelona_Goals']
    madrid_goals = row['Real_Madrid_Goals']
    recorded_winner = row['Winner']
    
    # Determine the actual winner based on goals
    if barcelona_goals > madrid_goals:
        actual_winner = 'Barcelona'
    elif madrid_goals > barcelona_goals:
        actual_winner = 'Real Madrid'
    else:
        actual_winner = 'Draw'
    
    # Check if recorded winner matches actual winner
    if recorded_winner != actual_winner:
        return {
            'Date': row.get('Date', 'N/A'),
            'Barcelona_Goals': barcelona_goals,
            'Real_Madrid_Goals': madrid_goals,
            'Recorded_Winner': recorded_winner,
            'Actual_Winner': actual_winner
        }
    return None


# Check each row and collect discrepancies
discrepancies = []
for index, row in df.iterrows():
    result = verify_winner(row)
    if result:
        discrepancies.append(result)

# Print results
if discrepancies:
    print("Found incorrect winner entries:")
    for disc in discrepancies:
        print(f"\nMatch details:")
        print(f"\nDate: {disc['Date']}")
        print(f"Barcelona {disc['Barcelona_Goals']} - {disc['Real_Madrid_Goals']} Real Madrid")
        print(f"Recorded winner: {disc['Recorded_Winner']}")
        print(f"Actual winner should be: {disc['Actual_Winner']}")
else:
    print("All winner entries are correct!")
```

```python
def split_scorers_data(df):
    # Create empty lists to store the expanded data
    expanded_data = []
    
    # Iterate through each row of the original dataframe
    for _, row in df.iterrows():
        # Split Barcelona scorers
        barca_scorers = str(row['Barcelona_Scorers']).split(', ') if pd.notna(row['Barcelona_Scorers']) else []
        # Split Real Madrid scorers
        madrid_scorers = str(row['Real_Madrid_Scorers']).split(', ') if pd.notna(row['Real_Madrid_Scorers']) else []
        
        # Get the maximum length between both scorer lists
        max_scorers = max(len(barca_scorers), len(madrid_scorers))
        
        # Pad the shorter list with None to match lengths
        barca_scorers.extend([None] * (max_scorers - len(barca_scorers)))
        madrid_scorers.extend([None] * (max_scorers - len(madrid_scorers)))
        
        # Create a record for each scorer
        for barca_scorer, madrid_scorer in zip(barca_scorers, madrid_scorers):
            expanded_data.append({
                'Date': row['Date'],
                'Venue': row['Venue'],
                'League': row['League'],
                'Barcelona_Goals': row['Barcelona_Goals'],
                'Real_Madrid_Goals': row['Real_Madrid_Goals'],
                'Winner': row['Winner'],
                'Barcelona_Scorer': barca_scorer,
                'Real_Madrid_Scorer': madrid_scorer
            })
    
    # Create new dataframe from the expanded data
    new_df = pd.DataFrame(expanded_data)
    return new_df


```

```python
new_df = split_scorers_data(df)
new_df.to_csv("extended.csv", index=False)
```
